## Environment Setup

```python
# Initial Imports
import os
import plaid
import requests
import datetime
import json
import pandas as pd
from dotenv import load_dotenv

%matplotlib inline
```

```python
# Helper functions

def pretty_print_response(response):
    """
    This function formats a JSON file to pretty printing it in the screen.
    """
    print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# Set environment variables from the .env file
load_dotenv()
```

```python
# Extract API keys from environment variables
PLAID_CLIENT_ID = os.getenv("PLAID_CLIENT_ID")
PLAID_SECRET = os.getenv("PLAID_SECRET")
PLAID_PUBLIC_KEY = os.getenv("PLAID_PUBLIC_KEY")
```

## Plaid Access Token

In this section, you will use the `plaid-python` API to generate the correct authentication tokens to access data in the free developer Sandbox. This mimics how you might connect to your own account or a customer account, but due to privacy issues, this homework will only require connecting to and analyzing the fake data from the developer sandbox that Plaid provides. 

Complete the following steps to generate an access token:

1. Create a client to connect to plaid

2. Use the client to generate a public token and request the following items: ['transactions', 'income', 'assets']

3. Exchange the public token for an access token

4. Test the access token by requesting and printing the available test accounts

### 1. Create a client to connect to plaid

```python
# Create the client object
client = plaid.Client(client_id=PLAID_CLIENT_ID, secret=PLAID_SECRET, public_key=PLAID_PUBLIC_KEY, environment='sandbox')
```

```python
# Set the institution id
INSITUTION_ID = "ins_109508"
```

### 2. Generate a public token

```python
# Create the public token
tkn_rspnse = client.Sandbox.public_token.create(INSITUTION_ID, ['transactions', 'income', 'assets'])
exch_rspnse = client.Item.public_token.exchange(tkn_rspnse['public_token'])

# Print the public token
print(tkn_rspnse)
```

### 3. Exchange the public token for an access token

```python
# Retrieve the access token
exch_rspnse = client.Item.public_token.exchange(tkn_rspnse['public_token'])
# Print the access toke
print(exch_rspnse)
```

```python
# Store the access token in a Python variable
access_token = exch_rspnse['access_token']
```

### 4. Fetch Accounts

```python
# Retrieve accounts
accounts = client.Accounts.get(access_token)   
# Print accounts
pretty_print_response(accounts)
```

# Account Transactions with Plaid

In this section, you will use the Plaid Python SDK to connect to the Developer Sandbox account and grab a list of transactions. You will need to complete the following steps:


1. Use the access token to fetch the transactions for the last 90 days

2. Print the categories for each transaction type

3. Create a new DataFrame using the following fields from the JSON transaction data: `date, name, amount, category`. (For categories with more than one label, just use the first category label in the list)

4. Convert the data types to the appropriate types (i.e. datetimeindex for the date and float for the amount)

### 1. Fetch the Transactions for the last 90 days

```python
# Set the start and end date 90 days from today
start_date = '{:%Y-%m-%d}'.format(datetime.datetime.now() + datetime.timedelta(-90))
end_date = '{:%Y-%m-%d}'.format(datetime.datetime.now())
# Retrieve the transactions for the last 90 days
transactions = client.Transactions.get(access_token, start_date, end_date)
# Print the transactions
pretty_print_response(transactions)
```

```python
transactions
```

### 2. Print the categories for each transaction

```python
# Create a for-loop to print the categories for each transaction
for transaction in transactions["transactions"]:
    print(transaction["category"])
```

```python
df1 = pd.DataFrame(columns=['date', 'name', 'amount', 'category'])
transactions1 = transactions['transactions']

for index, transaction in enumerate(transactions1):
#     print(index, transaction)
    df1.loc[index] = [transaction['date'], transaction['name'], transaction['amount'], transaction['category'][0]]
df1.head()
```

### 3. Create a new DataFrame using the following fields from the JSON transaction data: `date`, `name`, `amount`, `category`. 

(For categories with more than one label, just use the first category label in the list)

```python
# Define an empty DataFrame and set the columns as date, name, amount, and category
df = pd.DataFrame(columns=['date', 'name', 'amount', 'category'])

#dates = []

# Retrieve all the transactions
for transaction in transactions['transactions']:
# Populate the transactions DataFrame with the transactions data
    new_row = {'date':transaction["date"], 'name':transaction["merchant_name"], 'amount':transaction["amount"], 'category':transaction["category"][0]}
# Display sample data from the DataFrame  
    df = df.append(new_row, ignore_index=True)
    #print(new_row)
df.head()
```

### 4. Convert the data types to the appropriate types 

(i.e. datetimeindex for the date and float for the amount)

```python
# Display data types
df.dtypes
```

```python
# Make any required data type transformation
df["amount"] = df["amount"].astype(float)
df["date"] = pd.to_datetime(df["date"])
```

```python
# Set the date column as index
df.set_index(df['date'], inplace=True)
df.drop("date", inplace=True, axis=1)
# Display sample data
df.head()
```

### Calculate the expenses per category

```python
# Compute the total expenses per category
total_expenses = df.groupby('category').sum()['amount']
# Display the expenses per catefory
total_expenses
```

```python
# Create a spending categories pie chart.
total_expenses.plot(kind="pie")
```

### Calculate the expenses per month

```python
# Create a DataFrame with the total expenses
total_expense_month = df.groupby(df.index.month).sum()['amount']

# Display sample data
total_expense_month.head()
```
